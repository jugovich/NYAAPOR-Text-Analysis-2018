### Combine the batched datasets into one file

```python
# create container to store batched dataset, consider each dataset as a frame
frames = []
for x in range(0, 155000, 5000):
    # load the pickle file
    _df = pd.read_pickle(r'resources/data/IBMWatson_followers_%s.pkl' % x)
    # add the dataset to the list of frames
    frames.append(_df)

# concat all the frames into one dataset
full_df = pd.concat(frames)
# save the new dataset to a file
full_df.to_json(r'resources/data/IBMWatson_followers_entities.json')
```
